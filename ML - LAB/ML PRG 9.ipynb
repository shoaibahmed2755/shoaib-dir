```python
# Import required Libraries
from sklearn.datasets import fetch_olivetti_faces
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load the Olivetti Faces dataset
faces = fetch_olivetti_faces()
X = faces.data  # each face image as a flat feature vector
y = faces.target  # Labels: person IDs (0 to 39)
# X has shape (400, 4096) - 400 images, each 64 x 64 pixels = 4096 features.
# y has 40 classes (each person has 10 images).
# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
# 70% training, 30% test.
# random_state=42 makes results reproducible.
# Train the Naive Bayes classifier
model = GaussianNB()
model.fit(X_train, y_train)
# Naive Bayes assumes pixel values follow a Gaussian distribution per class.
# Predict on the test set
y_pred = model.predict(X_test)
# Predicts the class (person ID) for each face in the test set.
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
# Shows how well the model performs.
print("Naive Bayes Accuracy on Olivetti Faces:", accuracy)
for i in range(5):
    plt.imshow(X_test[i].reshape(64, 64), cmap='gray')
    plt.title(f"Actual: {y_test[i]}, Predicted: {y_pred[i]}")
    plt.axis('off')
    plt.show()
```
