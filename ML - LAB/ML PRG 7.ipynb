```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Load dataset
boston = pd.read_csv(r'C:\Users\IronCore-2024\.cache\kagglehub\datasets\altavish\boston-housing-dataset\versions\1\HousingData.csv')

# Feature: average number of rooms (RM)
X = boston[['RM']].values  # Corrected: Use square brackets to access column

# Extracts the column "RM" (average rooms per dwelling) from the dataset.
# values ensures it's a NumPy array. It's already 2D with double brackets.

# Target: median house value (in $1000s)
y = boston['MEDV'].values

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)  # Corrected: Use X instead of x

# Splits data: 80% for training and 20% for testing.
# random_state=0 ensures the results are reproducible.

# Train model
model = LinearRegression()
model.fit(X_train, y_train)

# Creates a linear regression model and fits it to the training data.

# Predict
y_pred = model.predict(X_test)

# Plots the actual test data points as red dots.
plt.plot(X_test, y_pred, color='blue', label='Predicted')

# Draws the predicted regression line in blue over the test data.
plt.xlabel('Average rooms per house (RM)')
plt.ylabel('House price ($1000s)')
plt.title('Linear Regression - Boston Housing')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split

# Load Auto MPG dataset from URL
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv"
data = pd.read_csv(url)

# Drop rows with missing values
data = data.dropna()

# Use 'horsepower' to predict 'mpg'
X = data[['horsepower']].values  # Ensure 2D array for sklearn
y = data['mpg'].values

# Polynomial transformation (degree 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.2, random_state=0)

# Train model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict
X_test_horsepower = X_test[:, 1]  # original horsepower values
y_pred = model.predict(X_test)

# Plot
plt.scatter(X_test_horsepower, y_test, color='green', label='Actual')
plt.scatter(X_test_horsepower, y_pred, color='purple', label='Predicted', alpha=0.6)
plt.xlabel('Horsepower')
plt.ylabel('Miles per Gallon (MPG)')
plt.title('Polynomial Regression - Auto MPG')
plt.legend()
plt.grid(True)
plt.show()
```
